```python
movies = movies[['id', 'title', 'genre', 'overview']]
```

```python
movies
```

```python
movies['tags'] = movies['overview']+movies['genre']
```

```python
movies
```

```python
new_movie = movies.drop(columns = ['overview','genre'])
```

```python
new_movie
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=10000,stop_words = 'english')
```

```python
cv
```

```python
vec = cv.fit_transform(new_movie['tags'].values.astype('U')).toarray()
```

```python
vec
```

```python
vec.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sm = cosine_similarity(vec)
```

```python
sm
```

```python
new_movie[new_movie['title'] == "The Godfather"]
```

```python
new_movie[new_movie['title'] == "The Godfather"].index[0]
```

```python
new_movie.info()
```

```python
dis = sorted(list(enumerate(sm[2])),reverse = True, key = lambda vec : vec[1])
```

```python
for i in dis[0:5]:
  print(new_movie.iloc[i[0]].title)
```

```python
def recommend(movies):
  index = new_movie[new_movie['title'] == movies].index[0]
  dis = sorted(list(enumerate(sm[index])),reverse = True, key = lambda vec : vec[1])
  for i in dis[0:5]:
    print(new_movie.iloc[i[0]].title)
```

```python
recommend(input("Enter your Favourite  movie name : "))
```
